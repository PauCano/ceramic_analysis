```python
import os
import path
import glob
import random
import shutil

maskFolder = "./NN_DatasetV3/masks/"
imageFolder = "./NN_DatasetV3/images/"
trainFolder = './NN_DatasetV3/Training/'
validationFolder = './NN_DatasetV3/Validation/'
testFolder = './NN_DatasetV3/Test/'
```

```python
folders = [trainFolder, validationFolder, testFolder]
#files = os.listdir(maskFolder)
files = glob.glob(maskFolder+"g[1-6]*")
files = [file.split("\\")[-1] for file in files]
files = [file.split(".")[0] for file in files]
random.shuffle(files)
print(files)
setTotalImages = len(files)
percentile60 = int(setTotalImages*0.6)
percentile90 = int(setTotalImages*0.9)

for i in range(len(folders)):
    bottom = 0
    top = setTotalImages
    folder = folders[i]
    print("Processing folder: "+folder+"...")
    subfolders = ["images/", "masks/"]
    for subfolder in subfolders:
        folderFiles = glob.glob(folder+subfolder+"*")
        for f in folderFiles:
            f = f.replace("\\", "/")
            os.stat(f)
            os.remove(f)
    if i==0:
        top = percentile60
    elif i==1:
        bottom = percentile60
        top = percentile90
    else:
        bottom = percentile90
    '''os.mkdir(folder)
    os.mkdir(folder+"masks/")
    os.mkdir(folder+"images/")'''
    
    for el in range(bottom, top):
        print("File "+str(el)+" of "+str(bottom)+" to "+str(top))
        shutil.copyfile(maskFolder + files[el]+".png", folder+"masks/"+files[el]+".png")
        shutil.copyfile(imageFolder + files[el]+".hdr", folder+"images/"+files[el]+".hdr")
        shutil.copyfile(imageFolder + files[el]+".img", folder+"images/"+files[el]+".img")
        shutil.copyfile(imageFolder + files[el]+".png", folder+"images/"+files[el]+".png")
    

'''
x_train = x[0:percentile60]
x_val = x[percentile60:percentile90]
x_test = x[percentile90:]
x_t= x[:]

y_train = y[0:percentile60]
y_val = y[percentile60:percentile90]
y_test = y[percentile90:]
y_t = y[:]'''
```
